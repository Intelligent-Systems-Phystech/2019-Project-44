```python
name = 'servo'
```

```python
X, y = ut.DataLoader(name=name)
X, y = ut.preprocess(X, y)
```

```python
real_data = mm.ACC(X, y, progress_bar=lambda x: tqdm(x, leave = False))
```

```python
statmodel = ut.define_model(y)
```

```python
sample_size = np.arange(10, 500, 5)
print(X.shape)
```

```python
def iDistribution(cov, l=0.25):

    W = sps.multivariate_normal(mean=w_hat, cov = cov).rvs(size=10000)

    return (np.sqrt(((W - w_hat)**2).sum(axis = 1)) < l).mean()
```

```python
num_iters = 1000

r = 10

sample_sizes = np.arange(10, 3000, 5)  

plt.plot(real_data['m'][r:], real_data['E'][r:], lw = 1.5, label='реальное значение')
plt.title('Зависимость реального уровня доверия доверительного шара от объема выборки')

for m in [50, 100, 150, 200]:
    
    X_m, y_m = ut.get_subset(X, y, m, duplications=False)
    w_hat = ut.get_params(X_m, y_m)
    
    answers = []
    
    for i in tqdm(range(num_iters)):
        X_sub, y_sub = ut.get_subset(X_m, y_m, m // 2, duplications=False)
        model = statmodel(y_sub, X_sub)
        w_hat = ut.get_params(X_sub, y_sub)
        answers.append((-model.hessian_fixed(w_hat) / X_sub.shape[0]))
    
    answers = np.asarray(answers)
    
    i_matrix = np.linalg.inv(np.mean(answers, axis = 0))
      
    alphas = []

    for cur_m in tqdm(sample_sizes):
        alphas.append(iDistribution(i_matrix / cur_m))
    
    alphas = np.asarray(alphas)
    
    plt.plot(sample_sizes[r:], alphas[r:], lw = 0.8, label='предсказание при m = {}'.format(m))

plt.xlabel('объем выборки')
plt.ylabel('радиус шара')
plt.grid()
plt.legend()
plt.show()
```

```python
m = 50
r = 10

sample_sizes = np.arange(10, 3 * X.shape[0], 5)

many_alphas = []

plt.plot(real_data['m'][r:], real_data['E'][r:], label='реальное значение')
#plt.title('Зависимость реального уровня доверия доверительного шара от объема выборки')

for i in range(20):
    
    X_m, y_m = ut.get_subset(X, y, m, duplications=False)
    w_hat = ut.get_params(X_m, y_m)
    
    answers = []
    
    for i in range(num_iters):
        X_sub, y_sub = ut.get_subset(X_m, y_m, m // 2, duplications=False)
        model = statmodel(y_sub, X_sub)
        w_hat = ut.get_params(X_sub, y_sub)
        answers.append((-model.hessian_fixed(w_hat) / X_sub.shape[0]))
    
    answers = np.asarray(answers)
    
    i_matrix = np.linalg.inv(np.mean(answers, axis = 0))
    
    alphas = []

    for cur_m in sample_sizes:
        alphas.append(iDistribution(i_matrix / cur_m))
    
    alphas = np.asarray(alphas)
    many_alphas.append(alphas)

many_alphas = np.asarray(many_alphas)

alpha = 0.95

left_q = sps.norm(loc=many_alphas.mean(axis=0)[r:], scale=many_alphas.std(axis=0)[r:]).ppf((1 - alpha) / 2)
right_q = sps.norm(loc=many_alphas.mean(axis=0)[r:], scale=many_alphas.std(axis=0)[r:]).ppf(1 - (1 - alpha) / 2)

plt.plot(sample_sizes[r:], many_alphas.mean(axis=0)[r:], color='red', label='предсказание при m = {}'.format(m))
plt.fill_between(sample_sizes[r:], left_q,right_q, 
                 color='red', lw=0.7, alpha=0.1, label='95%-ый доверительный интервал')

plt.xlabel('объем выборки')
plt.ylabel('реальный уровень доверия')
plt.legend()
plt.grid()
plt.savefig('../data/pics/ACC_modified_{}.pdf'.format(name), format='pdf')
plt.show()
```
